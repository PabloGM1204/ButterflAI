# Datos del modelo

Este es un cuaderno que resumen un poco el modelo con el *fine-tunning* hecho:

```python
from ultralytics import YOLO
from torchinfo import summary
import torch

# Carga tu modelo YOLOv8 fine-tuneado
model = YOLO('D:/TFM/Codigo/butterfly_detection/yolov8_butterfly/weights/best.pt')

# Definir el tamaño de entrada (por ejemplo, 640x640)
input_size = (1, 3, 640, 640)

# Imprimir un resumen del modelo para ver su arquitectura y las dimensiones en cada capa
print("Resumen del modelo:")
print(summary(model.model, input_size=input_size))

# Realizar un pase hacia adelante con un tensor de entrada dummy para inspeccionar la salida
dummy_input = torch.randn(*input_size)
with torch.no_grad():
    outputs = model.model(dummy_input)

# Función recursiva para imprimir las formas de los tensores en la salida
def print_shapes(data, indent=0):
    prefix = " " * indent
    if isinstance(data, torch.Tensor):
        print(prefix + "Tensor shape:", data.shape)
    elif isinstance(data, (list, tuple)):
        for i, item in enumerate(data):
            print(prefix + f"Elemento {i}:")
            print_shapes(item, indent + 2)
    else:
        print(prefix + f"Tipo {type(data)} no tiene atributo 'shape'.")

print("\nTipo de salida y formas:")
print_shapes(outputs)
```
